```python
f= open('gcn_2sumpooling_epoch1000.txt','r')
f1 = open('gin_proteins_epoch500.txt', 'r')
f2 = open('hgpsl_bs64_gl_epoch100.txt','r')
f3 = open('SAG-accPROTEINS.txt','r')
f4 = open('newmodel_mse_bs64_epoch300.txt', 'r')
```

```python
data = f.readlines()
data1 = f1.readlines()
data2 = f2.readlines()
data3 = f3.readlines()
data4 = f4.readlines()
```

```python
for i in range(len(data)):
    data[i] = data[i].split(' ')
for i in range(len(data1)):
    data1[i] = data1[i].split(' ')
for i in range(len(data2)):    
    data2[i] = data2[i].split(' ')
for i in range(len(data3)):
    data3[i] = data3[i].split(' ')
for i in range(len(data4)):
    data4[i] = data4[i].split(' ')
```

```python
i=0
for i in range(len(data)):
    for j in range(len(data[i])):
        data[i][j] = float(data[i][j])
i=0
for i in range(len(data1)):
    for j in range(len(data1[i])):
        data1[i][j] = float(data1[i][j])
i=0
for i in range(len(data2)):
    for j in range(len(data2[i])):
        data2[i][j] = float(data2[i][j])
i=0
for i in range(len(data3)):
    for j in range(len(data3[i])):
        data3[i][j] = float(data3[i][j])
i=0
for i in range(len(data4)):
    for j in range(len(data4[i])):
        data4[i][j] = float(data4[i][j])
```

```python
avg_loss=[]
avg_lossgin=[]
avg_losshgp=[]
avg_losssag=[]
avg_lossgan=[]
acc_train=[]
acc_test=[]
acc_testgin=[]
acc_testhgp=[]
acc_testsag=[]
acc_testgan=[]
for d in data:
    avg_loss.append(d[0])
    acc_train.append(d[1])
    acc_test.append(d[2])
for d1 in data1:
    acc_testgin.append(d1[2])
    avg_lossgin.append(d1[0])
for d2 in data2:
    acc_testhgp.append(d2[2])
    avg_losshgp.append(d2[0])
for d3 in data3:
    acc_testsag.append(d3[1])
    avg_losssag.append(d3[0])
for d4 in data4:
    acc_testgan.append(d4[2])
    avg_lossgan.append(d4[0])  
```
